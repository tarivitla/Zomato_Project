#### Q5. Create a seaborn plot to visualize the distribution of cousine types.
#### Which Seaborn function is suitable, and what insights can be gained? 

```python
sns.countplot(x='Cuisines',data=Dataset,order=pd.value_counts(Dataset['Cuisines']).iloc[:10].index, hue="Has Online delivery")
plt.xticks(rotation = 90)
```

#### In this countplot I found the top 10 cuisines in our dataset with the help of total counts.

```python
# Dataset['Cuisines'].value_counts()[:10].plot(kind='barh')

df1 = Dataset['Cuisines'].value_counts()[:10].reset_index()
```
